***Converting "Years in current job" this column into float datatype and***  ***converting values of that column into proper meaning ***

```python
df["Years in current job"].unique()
```

```python
# In my point of view i want this column in a float datatype 
# So for that I'm converting all the values into float datatype ,
# But Whenver there is <,>,+ this signs I will convert this into some another no.
# Like suppose if 10+ is a value so to detrmine it in a float,i will replpace it with 10.5 
# to take into consideration that this particular customer has 10 + years in current job 
# and for < 1 year i will put 0.5 
# to take into consideration that this particular customer has less than 1 years in current job 
```

```python
def replacee(year):
  if year == '10+ years':
    return '10.5'
  elif year == '< 1 year':
    return '0.5'
  else:
    return year
```

```python
df['Years in current job'].unique()
```

```python
df['Years in current job']
```

```python
df['Years in current job']=df['Years in current job'].apply(lambda x:replacee(x))
```

```python
df["Years in current job"].str.replace('years','')
```
